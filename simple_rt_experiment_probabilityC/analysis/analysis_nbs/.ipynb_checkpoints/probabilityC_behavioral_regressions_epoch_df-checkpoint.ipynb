```r
library(corrplot)
# mat : is a matrix of data
# ... : further arguments to pass to the native R cor.test function
cor.mtest <- function(mat, ...) {
    mat <- as.matrix(mat)
    n <- ncol(mat)
    p.mat<- matrix(NA, n, n)
    diag(p.mat) <- 0
    for (i in 1:(n - 1)) {
        for (j in (i + 1):n) {
            tmp <- cor.test(mat[, i], mat[, j], ...)
            p.mat[i, j] <- p.mat[j, i] <- tmp$p.value
        }
    }
  colnames(p.mat) <- rownames(p.mat) <- colnames(mat)
  p.mat
}
# matrix of the p-value of the correlation
p.mat <- cor.mtest(epoch_ls_df_reg_contin)

corrplot(epoch_ls_df_reg_contin, method="number", type="upper", order="hclust", 
         p.mat = p.mat, sig.level = 0.05, insig = "blank")
```

```r
#create copy of df for prediction & simulation
vol_epoch_df_reg_sim = vol_epoch_df_reg
vol_epoch_df_reg_pred_sim = vol_epoch_df_reg

#create new subject ids 
vol_epoch_df_reg_sim$ID = vol_epoch_df_reg$ID + n_subjects
vol_epoch_df_reg_sim$ID = as.factor(vol_epoch_df_reg_sim$ID)

vol_epoch_df_reg_pred_sim$ID_num = vol_epoch_df_reg$ID + n_subjects

head(vol_epoch_df_reg_sim)

vol_epoch_df_reg_pred_sim$p_id_accuracy <- ifelse(vol_epoch_df_reg_pred_sim$ID %in% vol_epoch_df_reg_sim$ID, NA, vol_epoch_df_reg_pred_sim$p_id_accuracy)
vol_epoch_df_reg_pred_sim$rt <- ifelse(vol_epoch_df_reg_pred_sim$ID %in% vol_epoch_df_reg_sim$ID, NA, vol_epoch_df_reg_pred_sim$rt)
```

```r
unique(vol_epoch_df_reg_sim$ID_num)
```

```r
vol_epoch_df_reg_pred_sim$simulated_acc <- simulate(vol_acc_reg, seed=1, newdata=vol_epoch_df_reg_pred_sim, re.form=NA,
                        allow.new.levels=T)$sim_1
vol_epoch_df_reg_pred_sim$predicted_acc <- predict(vol_acc_reg, newdata=vol_epoch_df_reg_pred_sim, allow.new.levels=T)
```

```r
vol_epoch_df_reg_pred_sim$simulated_rt <- simulate(vol_rt_reg, seed=1, newdata=vol_epoch_df_reg_pred_sim, re.form=NA,
                        allow.new.levels=T)$sim_1
vol_epoch_df_reg_pred_sim$predicted_rt <- predict(vol_rt_reg, newdata=vol_epoch_df_reg_pred_sim, allow.new.levels=T)
```

```r
acc_f <- function(x, ...) {
    plot(x, xlab="Subject", ylim=c(0, 1), ...)
    grid()
}
par(mfrow=c(1,3), mar=c(4,4,1,1))
with(vol_epoch_df_reg_pred_sim, acc_f(tapply(p_id_accuracy,  ID, mean), main="Original data", ylab="Acc", col=rep(1:2, each=24)))
with(vol_epoch_df_reg_pred_sim, acc_f(tapply(predicted_acc, ID, mean), main="Predicted data", ylab=""), col=rep(1:2, each=24))
with(vol_epoch_df_reg_pred_sim, acc_f(tapply(simulated_acc, ID, mean), main="Simulated data", ylab=""), col=rep(1:2, each=24))
legend("bottomright", pch=c(1,1), col=1:2, c("old subjects", "new subjects"), bg="white")
```
