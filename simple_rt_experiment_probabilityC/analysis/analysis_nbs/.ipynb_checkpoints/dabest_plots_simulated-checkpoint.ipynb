```python
import dabest
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = '/Users/67981492/Dropbox/volatileValues/simple_rt_experiment_probabilityC/analysis/aggregated_data/'
ls_obs_data = pd.read_csv(data_path+"combined_learning_sig_obs_data.csv")
```

```python
ls_obs_data.head()
```

```python
grouped_df = ls_obs_data.groupby(['ID', 'condition'])[['p_id_accuracy', 'rt']].mean()
grouped_df.head()
```

```python
n_boot=5000
```

```python
dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='p_id_accuracy', 
            color_col='condition', idx=['lc', 'hc'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Accuracy',  swarm_ylim = (0.4, 1))
plt.axhline(linewidth=2, color='k', y = -0.22, ls='dashed', alpha=0.3)

plt.
```

```python
dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='p_id_accuracy', 
            color_col='condition', idx=['lv', 'hv'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Accuracy', swarm_ylim = (0.4, 1))
plt.axhline(linewidth=2, color='k', y = -0.175, ls='dashed', alpha=0.3)
```

```python
dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='rt', 
            color_col='condition', idx=['lv', 'hv'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Reaction Time (s)', swarm_ylim = (0.24, 0.5))
```
