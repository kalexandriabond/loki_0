```python
all_sim_data.columns = all_sim_data.columns.str.strip()
all_sim_data.head()
```

```python
all_sim_data = all_sim_data.rename(index=str, columns={"acc": "response"})
all_sim_data = all_sim_data[['response', 'rt', 'cond', 'subj_idx', 'ideal_B', 'cpp']] 
all_sim_data = all_sim_data.rename(index=str, columns={'ideal_B': 'B'})
all_sim_data = all_sim_data[all_sim_data.rt<0.6]
all_sim_data.reset_index(drop=True, inplace=True)
```

```python
all_sim_data['cond'] = all_sim_data['cond'].str.strip()
all_sim_data.head()
```

```python
np.unique(all_sim_data.subj_idx), np.unique(all_sim_data.cond)
```

```python
all_sim_data = hddm.utils.flip_errors(all_sim_data)
```

```python
#learning signals
learning_signals =['cpp:C(cond)','B:C(cond)']
B = learning_signals[1]
cpp = learning_signals[0]
#targets
targets  = [ 'v ~ ', 'a ~ ', 't ~ ']
include=['v', 't', 'a', 'z']
v = targets[0]
a = targets[1]
tr = targets[2]

n_samples, n_burned_samples = 1000, 50
```

```python
v+B, a+cpp
```

```python
v_B_cpp_a_subj_obs_reg = hddm.HDDMRegressor(all_sim_data,[v+B, a+cpp],bias=False, include=include, informative=True)
v_B_cpp_a_subj_obs_reg.sample(n_samples,burn=n_burned_samples,dbname='v_B_cpp_a_subj_sim_reg_idealB.db', db='pickle')
```

```python
v_cpp_a_B_subj_obs_reg = hddm.HDDMRegressor(all_sim_data,[v+cpp, a+B],bias=False, include=include, informative=True)
v_cpp_a_B_subj_obs_reg.sample(n_samples,burn=n_burned_samples,dbname='v_cpp_a_B_subj_sim_reg_idealB.db', db='pickle')
```

```python
# v_B_cpp_a_subj_obs_reg.save('v_B_cpp_a_subj_sim_reg_TEST_choiceB')
# v_cpp_a_B_subj_obs_reg.save('v_cpp_a_B_subj_sim_reg_TEST_choiceB')
```

```python
dics = [v_B_cpp_a_subj_obs_reg.dic, v_cpp_a_B_subj_obs_reg.dic]
m_names = ['v_B & a_cpp', 'v_cpp & a_B']
```
