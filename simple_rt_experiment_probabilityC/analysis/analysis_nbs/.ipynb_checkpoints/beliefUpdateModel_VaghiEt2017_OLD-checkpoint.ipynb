```python
from scipy.stats.distributions import uniform, norm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot

df = pd.read_csv("highRE_lowC.csv")
columns = df.columns
df = df.rename(columns=dict(zip(columns, ['rA', 'rB', 'ssd', 'cp', 'rdelta', 'mu', 'sigma'])))
```

```python
def bayesian_belief(x,H,sN,low,up):
    """ 
    Arguments:
      - x     = vector with values to estimate mean
      - H     = hazard rate
      - sN    = std of normal distribution (generative sigma)
      - low   = lower bound of uniform distribution (mu)
      - up    = upper bound of uniform distribution (mu)
    Returns:
      - B     = beliefs about underlying distribution (= estimate)
      - lr    = learning rate
      - d     = prediction error between current belief and observation
      - CPP   = change point probability
      - MC    = model confidence (= 1-RU)
      - r     = expected run length
      - sF    = expected amount of noise from the gen distribution
    Fabrice Luyckx, 20/3/2016
    """  
    # Get dimensions
    # sz = x.shape
    ttrials = x.size
    # nsamp = sz[1]
    # Initialise variables
    B = np.zeros(ttrials) # belief about current mean
    lr = B.copy() # learning rate
    d  = B.copy() # prediction error
    CPP = B.copy() # change-point probability (omega)
    MC = B.copy() + 0.5 # relative uncertainty (phi)
    r  = B.copy() + 1 # expected run length
    sF = B.copy() # expected amount of noise from the gen distribution
    high = up-low
    unifpdf = lambda x_t: uniform(low, high).pdf(x_t)
    normpdf = lambda mu, sig, x_t: norm(mu, sig).pdf(x_t)
    for t in range(0, ttrials):
        # Prediction error (observed point - belief)
        d[t] = x[t] - B[t]
        # Change-point probability
        # sF[t] = np.sqrt(sN**2 + (sN**2)/r[t]); # 2010 Paper
        sF[t] = np.sqrt(sN**2 + (sN**2 * (1-MC[t]))/(MC[t])) # 2012 Paper
        # CPP[t] = ((x[t],low,up)*H)/((unifpdf(x[t])*H) + (normpdf(diffcirc(x[t],B[t]),0,sF[t])*(1-H)))
        # CPP[t] = (unifpdf(x[t])*H)/((unifpdf(x[t])*H) + (normpdf(B[t]-x[t],sF[t],x[t])*(1-H)))
        CPP[t] = (unifpdf(x[t])*H)/((unifpdf(x[t])*H) + (normpdf(B[t],sF[t],x[t])*(1-H)))
        #KB: uniform dist. is modified to reflect the range of values for our task
        
        # Learning rate
        lr[t] = CPP[t] + (1-MC[t])*(1-CPP[t])
        # Next trial calculations
        if t < (ttrials-1):
            # Belief on next trial
            B[t+1] = B[t] + lr[t]*d[t]
            # MC on next trial
            # weighted average of the variance on the distribution conditional on a change point
            term1 = CPP[t]*sN**2 
            # weighted average of the variance on the distribution conditional on NO change point
            term2 = (1-CPP[t])*(1-MC[t])*sN**2
            # variance emerging from the difference in the means of two conditional distributions
            term3 = CPP[t]*(1-CPP[t])*(d[t]*(MC[t]))**2
            # denominator contains an additional term to account for uncertainty arising from noise
            MC[t+1] = 1 - ((term1+term2+term3)/(term1+term2+term3 + sN**2)) 
            # Expected run length on next trial
            r[t+1] = (r[t] + 1)*(1-CPP[t]) + CPP[t]        
    return [B,lr,d,CPP,MC,r,sF]
```

```python
x = df.rdelta.values
H = df[df.cp==1].shape[0] / df.shape[0]
sN = df.sigma.mean()
low = x.min()
up = x.max()
```

```python
out = bayesian_belief(x, H, sN, low, up)
B,lr,d,CPP,MC,r,sF = out
```

```python
jtplot.style(grid=False, fscale=1.5, figsize=(12,9))
```

```python
trials = np.arange(x.size)
plt.scatter(trials, x, label='$r_t$', color='#8172b2', alpha=.5)
plt.plot(trials, B, label='$\\mu_t$', color='#3572C6', linewidth=2)

#this is for a low reliability/volatility/conflict condition 
plt.title('Mean reward difference and reward difference observed')
plt.legend(fontsize=18)
ax = plt.gca()
ax.set_xlabel('Trials (t)')
sns.despine()
plt.show()
```

```python
upper = trials[-1]
plt.plot(MC[:upper],  linewidth=2.5, label='$\\phi$')
plt.plot(CPP[:upper],  linewidth=2.5, label='$\\Omega$')

cpdf = df.loc[:upper, :]
cpix = cpdf[df.cp==1].index.values
for cp_i in cpix:
    label=None
    if cp_i == cpix[-1]:
        label='State Change'
    plt.vlines(cp_i, 0, 1, color='k', linestyle='--', linewidth=2.5, alpha=.5, label=label)
plt.title('Bayesian model confidence and estimate of CPP')    
ax = plt.gca()
ax.set_ylim(0, 1)
ax.set_xlabel('Trials (t)')
plt.legend(loc=2, fontsize=18)
sns.despine()
plt.show()
```
