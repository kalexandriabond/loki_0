```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scikits.bootstrap as bootstrap  
```

```python
s = np.random.normal(3, 1, 600)
ci = scipy.stats.norm.interval(0.95, loc=3, scale=1)
print(ci)
```

```python
sns.distplot(s, color = 'black')
plt.xlabel('reward value')
plt.ylabel('probability')
plt.axvline(linewidth=1, color='k',x=ci[0], alpha=0.95,linestyle='--')
plt.axvline(linewidth=1, color='k',x=ci[1], alpha=.95,linestyle='--')
plt.savefig('reward_dist.png', dpi=600)
```
