```r
library(tidyverse)
library(readr)
library(ggplot2)

source("../../../spydr_scripts/functions/get_os.R")
```

```r
os <- get_os()
if (os == 'linux'){home <- '/home/krista/'
} else if(os == 'osx'){home <- '/Users/krista/'}

data_path = paste0(home,
                  'Dropbox/loki_0/simple_rt_experiment_probabilityC/data/')
setwd(data_path)
```

```r
(files <- dir(path = data_path, pattern = "^[0-9](.*)(_runInfo.csv)$"))
```

```r
metadata <- files %>%
  map(read_csv,col_types = cols()) %>%                  
  reduce(rbind)      
```

```r
(condition = map(files, substr, start=10, stop=10))
metadata$condition = as.numeric(condition)
```

```r
#condition codes from experimental script
# if condition == 0:
#         exp_param_file = exp_param_directory + 'lc_' + trialfile_n + '.csv'
#     elif condition == 1:
#         exp_param_file = exp_param_directory + 'hc_' + trialfile_n + '.csv'
#     elif condition == 2:
#         exp_param_file = exp_param_directory + 'hv_' + trialfile_n + '.csv'
#     elif condition == 3:
#         exp_param_file = exp_param_directory + 'lv_' + trialfile_n + '.csv'

#decode conditions
metadata$decoded_condition <- recode(metadata$condition,
  '0' = 'lc',
  '1' = 'hc',
  '2' = 'hv',
  '3' = 'lv', 
  )
#check decoding
metadata$decoded_condition[0:5]; metadata$condition[0:5]
```
