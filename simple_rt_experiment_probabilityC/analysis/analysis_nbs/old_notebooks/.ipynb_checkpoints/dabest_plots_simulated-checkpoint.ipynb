```python
import dabest
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
home_path = os.path.expanduser('~')
sim_data_path = home_path + '/Dropbox/loki_0/spydr_scripts/sim_obj_pkl/'
```

```python
ls_sim_data = pd.read_csv(sim_data_path+"simulated_behavior.csv")
```

```python
ls_sim_data.head()
ls_sim_data.loc[ls_sim_data.acc == 1, 'gen_drift'] = abs(ls_sim_data.loc[ls_sim_data.acc == 1, 'gen_drift']) 
# ls_sim_data = ls_sim_data.loc[ls_sim_data.rt < 0.6]
```

```python
ls_sim_data.loc[ls_sim_data.cond == 'hc', 'cond'] = 'high conflict'
ls_sim_data.loc[ls_sim_data.cond == 'lc', 'cond'] = 'low conflict'

ls_sim_data.loc[ls_sim_data.cond == 'lv', 'cond'] = 'low volatility'
ls_sim_data.loc[ls_sim_data.cond == 'hv', 'cond'] = 'high volatility'
```

```python
grouped_df = ls_sim_data.groupby(['ID', 'cond'], as_index=False)[['acc', 'rt', 'gen_bound', 'gen_drift']].mean()
grouped_df.head()
```

```python
rt_df = grouped_df.copy()
rt_df=rt_df.drop(['acc', 'gen_bound', 'gen_drift'], axis=1)

acc_df = grouped_df.copy()
acc_df=acc_df.drop(['rt','gen_bound', 'gen_drift'],axis=1)
acc_df=acc_df.pivot(index='ID', columns='cond',)
acc_df.head()
```

```python
gen_bound_df = grouped_df.copy()
gen_bound_df=gen_bound_df.drop(['acc', 'rt', 'gen_drift'], axis=1)
gen_bound_df=gen_bound_df.pivot(index='ID', columns='cond',)

gen_drift_df = grouped_df.copy()
gen_drift_df=gen_drift_df.drop(['rt','acc', 'gen_bound'],axis=1)
gen_drift_df=gen_drift_df.pivot(index='ID', columns='cond',)


gen_drift_df.columns = gen_drift_df.columns.droplevel()
gen_bound_df.columns = gen_bound_df.columns.droplevel()

gen_drift_df=gen_drift_df.reset_index(drop=True)
gen_bound_df=gen_bound_df.reset_index(drop=True)
```

```python
rt_df=rt_df.pivot(index='ID', columns='cond',)
rt_df.head()

rt_df.columns = rt_df.columns.droplevel()
acc_df.columns = acc_df.columns.droplevel()

rt_df=rt_df.reset_index(drop=True)
acc_df=acc_df.reset_index(drop=True)
```

```python
n_boot=5000
```

```python
f_conf_drift, results_conf_drift = dabest.plot(data=gen_drift_df, fig_size=(5,7),   
             idx=['low conflict', 'high conflict'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Simulated Drift-Rate, $v$',  swarm_ylim = (0.4, 0.9), custom_palette=['red','blue'])
plt.axhline(linewidth=2, color='k', y = -0.221, ls='dashed', alpha=0.3)
plt.savefig('sim_conflict_drift_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_drift, results_vol_drift = dabest.plot(data=gen_drift_df, fig_size=(5,7),  
           idx=['low volatility', 'high volatility'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Simulated Drift-Rate, $v$',  swarm_ylim = (0.4, 0.9),custom_palette=['darkorchid','forestgreen'])
plt.axhline(linewidth=2, color='k', y = -0.221, ls='dashed', alpha=0.3)
plt.savefig('sim_vol_drift_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_threshold, results_vol_threshold = dabest.plot(data=gen_bound_df, fig_size=(5,7), 
            idx=['low volatility', 'high volatility'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Simulated Decision Threshold, $a$',  swarm_ylim = (0.5, 0.7),custom_palette=['darkorchid','forestgreen'])
plt.axhline(linewidth=2, color='k', y = -0.221, ls='dashed', alpha=0.3)
plt.savefig('sim_vol_thresh_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_conf_threshold, results_conf_threshold = dabest.plot(data=gen_bound_df, fig_size=(5,7),  
             idx=['low conflict', 'high conflict'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Simulated Decision Threshold, $a$',  swarm_ylim = (0.5, 0.7),custom_palette=['red','blue'])
plt.axhline(linewidth=2, color='k', y = -0.221, ls='dashed', alpha=0.3)
plt.savefig('sim_conf_thresh_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_conf_acc, results_conf_acc = dabest.plot(data=acc_df, fig_size=(5,7),  
            idx=['low conflict', 'high conflict'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Simulated Accuracy',  swarm_ylim = (0.4, 1),
                                           custom_palette=['red','blue'])
plt.axhline(linewidth=2, color='k', y = -0.234, ls='dashed', alpha=0.3)
plt.savefig('sim_conf_acc_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_acc, results_vol_acc =dabest.plot(data=acc_df, fig_size=(5,7), 
            idx=['low volatility', 'high volatility'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Simulated Accuracy', swarm_ylim = (0.4, 1),
                                        custom_palette=['darkorchid','forestgreen'])
plt.axhline(linewidth=2, color='k', y = -0.228, ls='dashed', alpha=0.3)
plt.savefig('sim_vol_acc_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_rt, results_vol_rt = dabest.plot(data=rt_df, fig_size=(5,7), 
             idx=['low volatility', 'high volatility'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Simulated Reaction Time (s)', swarm_ylim = (0.24, 0.5),
                                       custom_palette=['darkorchid','forestgreen'])
plt.savefig('sim_vol_rt_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```
