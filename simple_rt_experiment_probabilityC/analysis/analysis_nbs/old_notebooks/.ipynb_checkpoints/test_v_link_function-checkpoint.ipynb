```python
#define v link function 
#this will flip the sign of v based on the stimulus column

#this is not what i want
#just look at default (+/- drift terminating on lower/upper bounds)
#this is in line with +/- belief. 
def v_link_func(x, data=all_sim_data):
    stim = (np.asarray(dmatrix('0 + C(s,[[1],[-1]])', {'s':data.stimulus.ix[x.index]})))
#     print('input', x)
#     print('flipped', x*stim)
    return x * stim
```

```python
def v_link_func2(v, data=all_sim_data):
    if data.stimulus != data.response:
        v = -abs(v) 
    else if data.stimulus == data.response:
        v = v 
    return v 
```

```python
#stim coded with target_B and link function = ? 
#if link function flips v when v is negative and going in the correct direction 
#then the drift rate coeff. should be + 

#is it that v_link func will make drift negative when going toward incorrect boundary?
test_stim_code_link_m = hddm.HDDMRegressor(all_sim_data,{'model':'v~B', 'link_func': v_link_func}, bias=False, include=['a', 'v', 'z', 't'], informative=True)
test_stim_code_link_m.sample(200,burn=50,dbname='test_stim_code_link.db', db='pickle')
```

```python
#is it that v_link func will make drift negative when going toward incorrect boundary?

#this one might be best, without a link -- when B is negative, v should be negative 
test_stim_code_no_link_m = hddm.HDDMRegressor(all_sim_data,{'model':'v~B', 'link_func': lambda x:x}, bias=False, include=['a', 'v', 'z', 't'], informative=True)
test_stim_code_no_link_m.sample(200,burn=50,dbname='test_stim_code_no_link.db', db='pickle')
```
